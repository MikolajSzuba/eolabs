# Lab 5 Quickstart

This notebook helps you verify that the Lab 5 viewer is ready to run.
It does not replace the GUI app in `viewer.py`.

```python
from pathlib import Path

data_dir = Path("data/images")
print(f"Data directory exists: {data_dir.exists()}")
if data_dir.exists():
    print(f"Files in {data_dir}:")
    for p in sorted(data_dir.iterdir()):
        print(" -", p.name)
```

```python
from pathlib import Path

hdr_files = sorted(Path("data/images").glob("*.hdr"))
print(f"Found {len(hdr_files)} header file(s).")
if hdr_files:
    print("Run one of these commands in a terminal:")
    print("  python viewer.py")
    print(f"  python viewer.py {hdr_files[0]}")
else:
    print("No .hdr files found. Copy your .hdr/.bsq pair into data/images first.")
```
